# Proving data classification algorithms

```python
from Cluster import Cluster #the class created (use help function for documentation)
from sklearn_extensions.fuzzy_kmeans import KMedians, FuzzyKMeans
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import make_blobs, make_circles, make_moons, make_s_curve
```

## With remarkable groups:

```python
#creating ramarkable dataset randomly:
    
X, y = make_blobs(
n_samples=200, n_features=2, centers=3, cluster_std=0.5, shuffle=True,random_state=0)

data_toy1 = pd.DataFrame(X)
data_toy1.columns = ['x1','x2']
```

```python
#--------------------clustering with k-means--------------------
k_means_cluster = Cluster(method = 'k_means', k = 3, iter = 20)
data_transformed = k_means_cluster.fit_transform(data = data_toy1)


plt.figure(figsize=(3,3))
sns.set(style="darkgrid")
sns.scatterplot(data = data_transformed, x = data_transformed.x1, y = data_transformed.x2,
                hue = 'cluster', palette='dark')
plt.title("K-Means")
```

```python
#--------------------clustering with GMM_EM--------------------
GMM_cluster = Cluster(method = 'GMM_EM', k = 3, iter = 20)
data_transformed = GMM_cluster.fit_transform(data = data_toy1)


plt.figure(figsize=(3,3))
sns.set(style="darkgrid")
sns.scatterplot(data = data_transformed, x = data_transformed.x1, y = data_transformed.x2,
                hue = 'cluster', palette='dark')
plt.title('GMM_EM')
```

```python
#--------------------clustering with Kernel K-Means--------------------
kk_means_cluster = Cluster(method = 'kk_means', k = 3, iter = 20, sigma=0.5)
data_transformed = kk_means_cluster.fit_transform(data = data_toy1)


plt.figure(figsize=(3,3))
sns.set(style="darkgrid")
sns.scatterplot(data = data_transformed, x = data_transformed.x1, y = data_transformed.x2,
                hue = 'cluster', palette='dark')
plt.title('Kernel K-Means')
```

```python
#--------------------clustering with Fuzzy K-Means--------------------
fuzzy_kmeans = FuzzyKMeans(k=3, m=2)

fuzzy_kmeans.fit(data_toy1)
data_transformed = data_toy1.copy()
data_transformed['cluster'] = fuzzy_kmeans.labels_

plt.figure(figsize=(3,3))
sns.set(style="darkgrid")
sns.scatterplot(data = data_transformed, x = data_transformed.x1, y = data_transformed.x2,
                hue = 'cluster', palette='dark')
plt.title('Fuzzy K-Means')
```
